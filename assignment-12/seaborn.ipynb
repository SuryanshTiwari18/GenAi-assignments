# Asignment-12: Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("retail_sales_dataset.csv")
df.columns = df.columns.str.strip()

print(df.head())
print(df.info())
```

## Task-1: Relational Plot


```python
sns.relplot(data=df,x="Quantity",y="Total Amount",hue="Product Category",kind="line")
plt.show()
```

```python
sns.relplot(data=df,x="Quantity",y="Total Amount",hue="Product Category",kind="scatter")
plt.show()
```

## Task-2: Line Plot as Scatter & Facet

```python
sns.lineplot(data=df,x="Quantity",y="Total Amount")
plt.title("Line Plot")
plt.show()
```

```python
sns.relplot(data=df,x="Quantity",y="Total Amount",kind="scatter")
plt.show()
```

```python
sns.relplot(data=df,x="Quantity",y="Total Amount",hue="Product Category",col="Gender")
plt.show()
```

## Task-3: Distribution Plots

```python
sns.histplot(df["Total Amount"])
plt.title("Histogram")
plt.show()
```

```python
sns.kdeplot(df["Total Amount"])
plt.title("KDE Plot")
plt.show()
```

```python
sns.rugplot(df["Total Amount"])
plt.title("Rug Plot")
plt.show()
```

```python
sns.histplot(df["Total Amount"],kde=True)
plt.title("Histogram + KDE")
plt.show()
```

## Task-4: Bivariate Distribution Plots


```python
sns.histplot(data=df,x="Quantity",y="Total Amount")
plt.title("Bivariate Histogram")
plt.show()
```

```python
sns.kdeplot(data=df,x="Quantity",y="Total Amount",fill=True)
plt.title("Bivariate KDE")
plt.show()
```

## Task-5: Matrix Plots

```python
sns.pairplot(df[["Age","Quantity","Price per Unit","Total Amount"]])
plt.show()
```

```python
corr = df[["Age","Quantity","Price per Unit","Total Amount"]].corr()

sns.heatmap(corr,annot=True)
plt.title("Correlation Matrix")
plt.show()
```

## Task-6: Categorical Plots

```python
sns.barplot(data=df,x="Product Category",y="Total Amount")
plt.title("Bar Plot")
plt.show()
```

```python
sns.boxplot(data=df,x="Product Category",y="Total Amount")
plt.title("Box Plot")
plt.show()
```

```python
sns.violinplot(data=df,x="Product Category",y="Total Amount")
plt.title("Violin Plot")
plt.show()
```

```python
sns.countplot(data=df,x="Product Category")
plt.title("Count Plot")
plt.show()
```

## Task-7: Regression Plots

```python
sns.regplot(data=df,x="Quantity",y="Total Amount")
plt.title("Regression Plot")
plt.show()
```

```python
sns.lmplot(data=df,x="Quantity",y="Total Amount",hue="Gender")
plt.show()
```

## Task-8: Multi-Plots & Figure-Level Plots

```python
g = sns.FacetGrid(df,col="Product Category")
g.map(plt.scatter,"Quantity","Total Amount")
plt.show()
```

```python
sns.relplot(data=df,x="Quantity",y="Total Amount",hue="Product Category",col="Gender")
plt.show()
```

```python
sns.catplot(data=df,x="Product Category",y="Total Amount",kind="bar")
plt.show()
```

```python
sns.displot(data=df,x="Total Amount",kde=True)
plt.show()
```
