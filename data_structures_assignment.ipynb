# Task 3: Product Pricing (Dictionaries)

```python
# 1. create a dictionary
price_dict = {
    "Laptop" : 80000,
    "Phone" : 50000,
    "Tablet" : 30000,
    "Headphones" : 2000,
    "Keyboard" : 1500,
    "Mouse" : 800
}

# 2. Add new product
price_dict["Monitor"] = 12000

# update price
price_dict["Laptop"] = 75000

# remove product safely
product_to_remove = "Tablet" 
if product_to_remove in price_dict:
    del price_dict[product_to_remove]
else:
    print("Product not found")

print("Updated price dictiondary: ", price_dict)

# 3. Average price
total = sum(price_dict.values())
avg_price = total / len(price_dict)
print("Average price: ", avg_price)

# extra: max and min priced products
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)

print("Max priced product: ", max_product, "with price", price_dict[max_product])
print("Min priced product: ", min_product, "with price", price_dict[min_product])
```

# Task 4: Combined Operations

```python
# using previous data
products = list(price_dict.keys())

categories = ["Electronics", "Electronics", "Electronics", "Audio", "Computer Accessories", "Computer Accessories"]

# 1. create catalog (product, price, category)
catalog = []
for i in range(len(products)):
    catalog.append((products[i], price_dict[products[i]], categories[i]))

print("Catalog: ", catalog)

# 2. category_to_product dictionary
category_to_product = {}
for product,price,category in catalog:
    if category not in category_to_product:
        category_to_product[category] = []
    category_to_product[category].append(product)

print("Category to products: ", category_to_product)

# 3. category with maximum products
length_dict = {k : len(v) for k,v in category_to_product.items()}
max_category = max(length_dict, key=length_dict.get)
print("Category with maximum products: ", max_category, "with", length_dict[max_category], "products")
```
