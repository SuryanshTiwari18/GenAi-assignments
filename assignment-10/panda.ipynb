## Task-1: Pandas Series Basic

```python
import pandas as pd 
marks = pd.Series([78, 85, 90, 66, 72])
print("Series values:")
print(marks.values)

print("\nIndex:")
print(marks.index)

print("\nData type:")
print(marks.dtype)

print("\nFirst element:")
print(marks.iloc[0])

print("\nLast two elements:")
print(marks.iloc[-2:])
```

## Task-2: Mathematical Operations on Series

```python
print("Add 5 grace marks:")
print(marks + 5)

print("\nSubtract 2 marks:")
print(marks - 2)

print("\nMultiply by 1.05:")
print(marks * 1.05)

print('\nDivide by 2:')
print(marks / 2)
```

## Task-3: Python Functionalities on Series

```python
print("Maximum marks:", marks.max())
print("Minimum marks:", marks.min())
print("Sum of marks:", marks.sum())
print("Mean marks:", marks.mean())

passed = marks.apply(lambda x: x >= 70)
print('\nPassed status for each student:')
print(passed)

print("\nCount of students passed:", passed.sum())
```

## Task-4: Create a DataFrame

```python
students = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'Marks': [78, 85, 90, 66, 72],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}

df = pd.DataFrame(students)
print("DataFrame:")
print(df)

print("\nFirst 3 rows:")
print(df.head(3))

print("\nLast 2 rows:")
print(df.tail(2))

print("\nShape:", df.shape)
print("Column names:", df.columns.tolist())
```

## Task-5: Important DataFrame Functions

```python
print("Info:")
print(df.info())

print("\nDescribe:")
print(df.describe())

print("\nHead:")
print(df.head())

```

```python
print('\nTail:')
print(df.tail())

sorted_df = df.sort_values(by='Marks', ascending=False)
print("\nSorted by Marks in descending order:")
print(sorted_df)

reset_sorted_df = sorted_df.reset_index(drop=True)
print("\nSorted DataFrame with reset index:")
print(reset_sorted_df)
```

## Task-6: Filtering & Conditional Selection

```python
print("Students who scored more than 75 marks:")
print(df[df['Marks'] > 75])

print("\nStudents belonging to Math subject:")
print(df[df['Subject'] == 'Math'])

print("\nStudents who scored more than average marks:")
print(df[df['Marks'] > df['Marks'].mean()])

print("\nStudents who failed (marks < 70):")
print(df[df['Marks'] < 70])
```

## Task-7: Grouping & Basic Analysis

```python
grouped = df.groupby('Subject')

print("\nAverage marks per subject:")
print(grouped['Marks'].mean())

print("\nCount of students per subject:")
print(grouped['Marks'].count())

print("\nMaximum marks per subject:")
print(grouped['Marks'].max())
```

## Task-8: Pandas Plotting (Simple Graphs)

```python
import matplotlib.pyplot as plt

df.plot(x='Name', y='Marks', kind='bar', legend=False, title='Student Name vs Marks')
plt.show()

df['Marks'].plot(kind='line', title='Marks Line Graph')
plt.show()

df['Marks'].plot(kind='hist', title='Histograms of Marks')
plt.show()
```

## Task-9: Mini Use Case: Sales Data Analysis

```python
sales = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue': [1200, 1500, 900, 2000, 1800]
}

sales_df = pd.DataFrame(sales)

print("Sales DataFrame:")
print(sales_df)

print("\nTotal revenue:", sales_df['Revenue'].sum())
print("Average daily revenue:", sales_df['Revenue'].mean())

max_revenue_day = sales_df.loc[sales_df['Revenue'].idxmax(), 'Day']
print("Day with maximum revenue:", max_revenue_day)

print("\nDays where revenue > average:")
print(sales_df[sales_df['Revenue'] > sales_df['Revenue'].mean()])

sales_df.plot(x='Day', y='Revenue', kind='line', legend=False, title='Revenue vs Day')
plt.show()
```
