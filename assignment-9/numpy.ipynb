## Task 1: Creating NumPy Arrays

```python
import numpy as np 

arr1 = np.arange(1, 11)
arr2 = np.arange(1, 10).reshape(3, 3)
arr3 = np.array([10, 20, 30, 40, 50])

print("arr1:",arr1)
print("Shape:", arr1.shape)
print("Data Type:", arr1.dtype)

print("\narr2:\n", arr2)
print("Shape:", arr2.shape)
print("Data Type:", arr2.dtype)

print("\narr3:", arr3)
print("Shape:", arr3.shape)
print("Data Type:", arr3.dtype)
```

## Task 2: Important Mathematical Operations

```python
A = np.array([10, 20, 30, 40])
B = np.array([1, 2, 3, 4])

print("A + B =", A + B)
print("A - B =", A - B)
print("A * B =", A * B)
print("A / B =", A / B)
print("A ** 2 =", A ** 2)

print("np.add(A, B) =", np.add(A, B))
print("np.subtract(A, B) =", np.subtract(A, B))
print("np.multiply(A, B) =", np.multiply(A, B))
print("np.divide(A, B) =", np.divide(A, B))
print("np.power(A, 2) =", np.power(A, 2))
```

## Task 3: Important NumPy Mathematical Formulas

```python
values = np.array([2, 4, 6, 8, 10])

print("values:", values)
print("Square root:", np.sqrt(values))
print("Exponential:", np.exp(values))
print("Natural log:", np.log(values))
print("Sum:", np.sum(values))
print("Cumulative sum:", np.cumsum(values))
```

## Task 4: Aggregation Operations

```python
data = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("data:\n", data)
print("Row-wise sum:", np.sum(data, axis=1))
print("Column-wise sum:", np.sum(data, axis=0))
print("Minimum value:", np.min(data))
print("Maximum value:", np.max(data))
print("Overall mean:", np.mean(data))
```

## Task 5: Statistical Operations

```python
marks = np.array([78, 85, 90, 66, 72, 88, 95, 60])

print("Marks:", marks)
print("Mean:", np.mean(marks))
print("Median:", np.median(marks))
print("Standard Deviation:", np.std(marks))
print("Variance:", np.var(marks))
print("Minimum:", np.min(marks))
print("Maximum:", np.max(marks))
print("Range:", np.max(marks) - np.min(marks))
```

## Task 6: Percentiles & Sorting

```python
sorted_marks = np.sort(marks)
print("Sorted Marks:", sorted_marks)
print("25th Percentile:", np.percentile(marks, 25))
print("50th Percentile (Median):", np.percentile(marks, 50))
print("75th Percentile:", np.percentile(marks, 75))

average_marks = np.mean(marks)
count_above_average = np.sum(marks > average_marks)
print("Average Marks:", average_marks)
print("Number of students above average:", count_above_average)
```

## Task 7: Mini Use Case - Sales Analysis

```python
sales = np.array([1200, 1500, 900, 2000, 1800, 1700, 1600])

print("Sales:", sales)
print("Total Sales:", np.sum(sales))
print("Average Daily Sales:", np.mean(sales))
print("Highest sales day:", np.argmax(sales))
print("Lowest sales day:", np.argmin(sales))
print("Standard Deviation of Sales:", np.std(sales))

above_average_days = sales > np.mean(sales)
print("Days with above average sales:", sales[above_average_days])
```
