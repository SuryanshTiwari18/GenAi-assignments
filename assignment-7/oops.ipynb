## Task 1: Basic Class & Object Creation

```python
class Product:
    def __init__(self,name,price,category):
        self.name = name
        self.price = price
        self.category = category
    
    def get_info(self):
        print(f"Product Name: {self.name}")
        print(f"Product Price: {self.price}")
        print(f"Product Category: {self.category}")
    
    def apply_discount(self, percent):
        discounted_price = self.price - (self.price * percent / 100)
        return discounted_price
    
product1 = Product("Laptop", 1000, "Electronics")
product2 = Product("Smartphone", 500, "Electronics")

product1.get_info()
discounted_price = product1.apply_discount(10)
print(f"Discounted Price: {discounted_price}")
print()
product2.get_info()
discounted_price = product2.apply_discount(20)
print(f"Discounted Price: {discounted_price}")
```

## Task 2: Constructor & Encapsulation

```python
class Product:
    def __init__(self,name,price,category):
        self.name = name
        self.__price = price
        self.category = category
    
    def get_info(self):
        print(f"Product Name: {self.name}")
        print(f"Product Price: {self.__price}")
        print(f"Product Category: {self.category}")
    
    def get_price(self):
        return self.__price
    
    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price
        else:
            print("Price must be positive.")
        
product3 = Product("Tablet", 300, "Electronics")
product3.get_info()
product3.set_price(250)
product3.get_info()
product3.set_price(-50)
```

## Task 3: Inheritance (Single-Level)

```python
class Product:
    def __init__(self,name,price,category):
        self.name = name
        self.price = price
        self.category = category
    
    def get_info(self):
        print(f"Product Name: {self.name}")
        print(f"Product Price: {self.price}")
        print(f"Product Category: {self.category}")
    
class ElectronicProduct(Product):
    def __init__(self,name,price,category,warranty_years):
        super().__init__(name, price, category)
        self.warranty_years = warranty_years
    
    def get_info(self):
        super().get_info()
        print(f"Warranty: {self.warranty_years} years")
    
electronic_item = ElectronicProduct("Laptop", 1000, "Electronics", 2)
electronic_item.get_info()
```

## Task 4: Polymorphism

```python
class Product:
    def __init__(self,name,price,category):
        self.name = name
        self.price = price
        self.category = category
    
    def get_info(self):
        print("Product:", self.name, "| Price:", self.price, "| Category:", self.category)
    
class Laptop(Product):
    def __init__(self,name,price,category,ram):
        super().__init__(name, price, category)
        self.ram = ram
    
    def get_info(self):
        print(f"Laptop -> Name: {self.name}, Price: {self.price}, RAM: {self.ram}")
    
class Mobile(Product):
    def __init__(self,name,price,category,camera):
        super().__init__(name, price, category)
        self.camera = camera
    
    def get_info(self):
        print(f"Mobile -> Name: {self.name}, Price: {self.price}, Camera: {self.camera}")
    
items = [
    Laptop("Gaming Laptop", 1500, "Electronics", "16GB"),
    Mobile("Smartphone", 800, "Electronics", "12MP")
]
for item in items:
    item.get_info()
```

## Task 5: Abstraction (Using Abstract Base Class)

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass

class CreditCardPayment(Payment):
    def process_payment(self, amount):
        print(f"Processing credit card payment of ${amount}")

class UPIPayment(Payment):
    def process_payment(self, amount):
        print(f"Processing UPI payment of ${amount}")
    
payments = [ CreditCardPayment(), UPIPayment()]
for payment in payments:
    payment.process_payment(2500)
```

## Task 6: Magic Methods & Operator Overloading

```python
class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def __str__(self):
        return f"Product(name={self.name}, price={self.price}, category={self.category})"

    def __add__(self, other):
        if isinstance(other, Product):
            return self.price + other.price
        return NotImplemented

product_a = Product("Headphones", 2000, "Electronics")
product_b = Product("Mouse", 1200, "Accessories")

print(product_a)
print(product_b)
print("Combined Price:", product_a + product_b)
```

## Task 7: Mini Project — Simple Inventory System (OOP Only)

```python
class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def __str__(self):
        return f"Product(name={self.name}, price={self.price}, category={self.category})"

    def __add__(self, other):
        if isinstance(other, Product):
            return self.price + other.price
        return NotImplemented

    def get_info(self):
        print(self)


class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, name):
        for product in self.products:
            if product.name == name:
                self.products.remove(product)
                print(f"Removed: {name}")
                return
        print(f"Product not found: {name}")

    def get_total_value(self):
        total = 0
        for product in self.products:
            total += product.price
        return total

    def show_all_products(self):
        for product in self.products:
            product.get_info()


class Store:
    def __init__(self, store_name):
        self.store_name = store_name
        self.inventory = Inventory()

    def add_new_product(self):
        name = input("Enter product name: ")
        price = int(input("Enter product price: "))
        category = input("Enter product category: ")
        product = Product(name, price, category)
        self.inventory.add_product(product)
        print("Product added successfully.")

    def show_summary(self):
        print("Store Name:", self.store_name)
        print("Total Items:", len(self.inventory.products))
        print("Total Value:", self.inventory.get_total_value())

store = Store("Tech World")

store.inventory.add_product(Product("Laptop", 70000, "Electronics"))
store.inventory.add_product(Product("Keyboard", 1500, "Accessories"))
store.inventory.add_product(Product("Monitor", 12000, "Electronics"))

store.show_summary()
store.inventory.show_all_products()

p1 = Product("Speaker", 3000, "Electronics")
p2 = Product("Mouse", 800, "Accessories")
print("Combined price of two products:", p1 + p2)
```
