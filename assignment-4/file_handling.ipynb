## Task 1: Write Sales Records to a File

```python
sales = [1200, 450, 980, 1500, 3000]
with open("sales_data.txt", "w") as file:
    for sale in sales:
        file.write(str(sale) + "\n")
        
with open("sales_data.txt", "r") as file:
    print(file.read())
```

## Task 2: Read File in Different Ways

```python
with open("sales_data.txt", "r") as file:
    print("Using read():")
    print(file.read())

with open("sales_data.txt", "r") as file:
    print("Using readline():")
    print(file.readline())

with open("sales_data.txt", "r") as file:
    lines = file.readlines()

sales_list = []
for line in lines:
    line = line.strip()
    if line != "":
        sales_list.append(int(line))
print("Using readlines() converted to list of integers:")
print(sales_list)
```

## Task 3: Append New Sales

```python
new_sales = [5000, 2500, 1700]

with open("sales_data.txt", "a") as file:
    for sale in new_sales:
        file.write(str(sale) + "\n")

with open("sales_data.txt", "r") as file:
    print("After appending new sales:")
    print(file.read())

with open("sales_data.txt", "r") as file:
    lines = file.readlines()

print("Total number of lines after appending new sales:", len(lines))
```

## Task 4: Generate Summary Report from File

```python
total_sales = 0
highest_sale = None
lowest_sale = None
count = 0


with open("sales_data.txt", "r") as file:
    for line in file:
        line = line.strip()
        if line == "":
            continue
        
        sale = int(line)
        total_sales += sale
        count += 1

        if highest_sale is None or sale > highest_sale:
            highest_sale = sale

        if lowest_sale is None or sale < lowest_sale:
            lowest_sale = sale

average_sales = total_sales / count if count > 0 else 0
print(f"Total Sales: {total_sales}")
print(f"Average Sales: {average_sales}")
print(f"Highest Sale: {highest_sale}")
print(f"Lowest Sale: {lowest_sale}")
```

## Task 7: Export Discounted Prices

```python
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "camera": 5000
}

discount_percent = int(input("Enter discount percentage:"))

with open("discount_report.txt", "w") as file:
    file.write("Product | Original Price | Discounted Price\n")

    total_items = 0
    total_discounted_price = 0

    for product, original_price in prices.items():
        discounted_price = original_price - (original_price * discount_percent / 100)
        discounted_price = round(discounted_price, 2)
        file.write(product + " | " + str(original_price) + " | " + str(discounted_price) + "\n")
        total_items += 1
        total_discounted_price += discounted_price
    
    average_discounted_price = round(total_discounted_price / total_items, 2) if total_items > 0 else 0
    file.write("\nTotal Items: " + str(total_items) + "\n")
    file.write("Average Discounted Price: " + str(average_discounted_price) + "\n")

with open("discount_report.txt", "r") as file:
    print("Contents of discount_report.txt:")
    print(file.read())
```
