## Task 1: Basic Function - Price After Discount

```python
def apply_discount(price, discount_percent = 5):
    if discount_percent > 60:
        discount_percent = 60
    discounted_price = price - (price * discount_percent / 100)
    return discounted_price

print(apply_discount(1000, 10))
print(apply_discount(500))
```

## Task 2: Recursive Function - Factorial Utility

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

## Task 3: Lambda Function - GST Calculator

```python
gst = lambda price: price + (price * 0.18)
print(gst(100))
```

## Task 4: Using map() - Apply GST to List of Prices

```python
prices = [100, 250, 400, 1200, 50]

gst = lambda price: price + (price * 0.18)

prices_with_gst = list(map(gst, prices))

print("Original prices:", prices)
print("Prices with GST:", prices_with_gst)
```

## Task 5: Using filter() - Filter Expensive Products

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]

greater_than_500 = list(filter(lambda price: price > 500, prices))
less_than_or_equal_500 = list(filter(lambda price: price <= 500, prices))

print("Prices greater than 500:", greater_than_500)
print("Prices less than or equal to 500:", less_than_or_equal_500)
```

## Task 6: Combined Utility Function

```python
def process_prices(prices):
    discounted_prices = list(map(lambda price: price - (price * 0.10), prices))
    filtered_prices = list(filter(lambda price: price > 300, discounted_prices))
    return discounted_prices, filtered_prices

discounted_prices, filtered_prices = process_prices([100, 500, 900, 50, 750])
print("Discounted prices:", discounted_prices)
print("Filtered prices:", filtered_prices)
```

## Task 7: Mini Problem - Menu Using Functions

```python
def add_price(prices_list, price):
    prices_list.append(price)
    
def get_average_price(prices_list):
    if len(prices_list) == 0:
        return 0
    total = 0
    for price in prices_list:
        total += price
    return total / len(prices_list)

def get_max_price(prices_list):
    if len(prices_list) == 0:
        return None
    max_price = prices_list[0]
    for price in prices_list:
        if price > max_price:
            max_price = price
    return max_price

prices = []
while True:
    print("\nMenu:")
    print("1 -> Add a price")
    print("2 -> Show average price")
    print("3 -> Show maximum price")
    print("q -> Exit")

    choice = input("Enter your choice: ").strip().lower()

    if choice == "q":
        print("Exiting the program.")
        break

    elif choice == "1":
        user_price = input("Enter a price to add: ")
        if user_price.isdigit():
            add_price(prices, int(user_price))
            print(f"Price {user_price} added.")
        else:
            print("Invalid input. Please enter a numeric price.")
    
    elif choice == "2":
        average_price = get_average_price(prices)
        print(f"Average price: {average_price}")
    
    elif choice == "3":
        max_price = get_max_price(prices)
        if max_price is not None:
            print(f"Maximum price: {max_price}")
        else:
            print("No prices available to determine maximum.")
    else:
        print("Invalid choice. Please select a valid option from the menu.")
```
