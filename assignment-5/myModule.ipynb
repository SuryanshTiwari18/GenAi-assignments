## Task 1: math_utils module


```python
import os
os.makedirs("modules_assignment", exist_ok=True)

with open("modules_assignment/math_utils.py", "w") as f:
    f.write("""
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def square(n):
    return n * n
""")

import sys
sys.path.append("modules_assignment")

import math_utils
from math_utils import square

print("Task 1 Output:")
print(math_utils.add(10, 5))
print(math_utils.subtract(10, 5))
print(square(6))
```

## Task 2: string_utils module

```python
with open("modules_assignment/string_utils.py", "w") as f:
    f.write("""
def capitalize_words(text):
    return " ".join(word.capitalize() for word in text.split())

def reverse_string(text):
    return text[::-1]

def word_count(text):
    return len(text.split())
""")
    
import string_utils
from string_utils import reverse_string

print("\nTask 2 Output:")
print(string_utils.capitalize_words("hello world"))
print(reverse_string("hello world"))
print(string_utils.word_count("hello world"))
```

## Task 3: Create Package (shop_package)

```python
os.makedirs("modules_assignment/shop_package", exist_ok=True)

with open("modules_assignment/shop_package/discount.py", "w") as f:
    f.write("""
def apply_discount(price, percent):
    return price - (price * percent / 100)

def flat_discount(price):
    return price - 50
""")

with open("modules_assignment/shop_package/billing.py", "w") as f:
    f.write("""
def calculate_total(prices):
    return sum(prices)

def apply_tax(amount):
    return amount + (amount * 5 / 100)
""")

with open("modules_assignment/shop_package/__init__.py", "w") as f:
    f.write("""
from .discount import apply_discount, flat_discount
from .billing import calculate_total, apply_tax
""")

print("\nTask 3: Package created successfully")
```

## Task 4: Import Package & Test

```python
import shop_package.discount as disc
from shop_package.billing import calculate_total
from shop_package import apply_tax, flat_discount, apply_discount

print("\nTask 4 Output:")

print(disc.apply_discount(1000, 10))
print(disc.flat_discount(1000))

print(calculate_total([100, 200, 300]))
print(apply_tax(1000))

print(flat_discount(1000))
print(apply_discount(1000, 15))
```
