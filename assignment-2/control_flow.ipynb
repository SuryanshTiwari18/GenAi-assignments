## Task - 1: Discount Rules(if/elif/else)

```python
user_input = input("Enter order amount: ")

if not user_input.isdigit():
    print("Invalid input. Please enter a numeric value.")
else:
    order_amount = int(user_input)
    if order_amount >= 2000:
        discount_percent = 15
    elif order_amount >= 1500:
        discount_percent = 10
    elif order_amount >= 1000:
        discount_percent = 7
    else:
        discount_percent = 0
    
    discount_amount = (order_amount * discount_percent) / 100
    final_amount = order_amount - discount_amount

    print("Order Amount: ", order_amount)
    print("Discount %: ", discount_percent)
    print("Discount Amount: ", discount_amount)
    print("Final Amount: ", final_amount)

    #optional tax part
    tax = (final_amount * 5) / 100
    total_amount = final_amount + tax
    print("Tax: ", tax)
    print("Total Amount: ", total_amount)
```

## Task - 2: Process Multiple Orders

```python
orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discounted_orders = 0

print("Order Amount -> Discount % -> Final Amount")
print("---------------------------------------------")

for order in orders:
    if order >= 2000:
        discount_percent = 15
    elif order >= 1500:
        discount_percent = 10
    elif order >= 1000:
        discount_percent = 7
    else:
        discount_percent = 0
    
    discount_amount = (order * discount_percent) / 100
    final_amount = order - discount_amount
    total_revenue += final_amount
    
    if discount_percent > 0:
        discounted_orders += 1
    
    print(f"{order} -> {discount_percent}% -> {final_amount}")

print("---------------------------------------------")
print(f"Total Revenue: {total_revenue}")
print(f"Number of Discounted Orders: {discounted_orders}")
```

## Task 3: User Menu

```python
orders = []

while True:
    print("\nMenu:")
    print("1 - Add order amount")
    print("2 - Show all orders and totals after discounts")
    print("q - quit")

    choice = input("Enter your choice: ").strip().lower()

    if choice == 'q':
        print("Exiting program...")
        break
    elif choice == '1':
        amount_input = input("Enter order amount: ")
        if not amount_input.isdigit():
            print("Invalid input. Please enter a numeric value.")
            continue
        
        orders.append(int(amount_input))
        print("Order added successfully.")
        continue
    elif choice == '2':
        if len(orders) == 0:
            print("No orders to display.")
            continue
        
        total_revenue = 0

        print("Order Amount -> Discount % -> Final Amount")
        print("---------------------------------------------")

        for order in orders:
            if order >= 2000:
                discount_percent = 15
            elif order >= 1500:
                discount_percent = 10
            elif order >= 1000:
                discount_percent = 7
            else:
                discount_percent = 0
            
            discount_amount = (order * discount_percent) / 100
            final_amount = order - discount_amount
            total_revenue += final_amount
            
            print(f"{order} -> {discount_percent}% -> {final_amount}")

        print(f"\nTotal Revenue: {total_revenue}")
    else:
        print("Invalid choice. Please select a valid option.")
        continue
```

## Task 4: Loop Control with Conditions

```python
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0

for sale in daily:
    if sale == -1:
        print("Corrupted data found. Stopping processing.")
        break
    if sale == 0:
        print("No sales for the day. Skipping.")
        continue

    total_sales += sale
    print(f"Added {sale} to total sales. Current total: {total_sales}")

print(f"Final Total Sales: {total_sales}")
```
